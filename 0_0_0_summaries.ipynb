# Die Stadt des Feuers

Die Handlung spielt in Trallop am Neunaugensee, einer Region voller Mythen, Naturgeister und der geheimnisvollen Fee Pandlaril. Der See gilt als magisch, launisch und manchmal gefährlich. In der Ferne ragt der schwarze Berg Cealan aus dem Wasser, von dem Rauch aufsteigt – ein unheilvolles Zeichen.

### Figuren

1. **Nindariel Pfeilgruß**

Ein Auelf aus den Pandlaril-Auen.
Er verkauft Holzschnitzereien, kennt die Natur und die Ufer des Sees sehr gut.
Er glaubt fest an Pandlaril und ihre schützende Rolle.

2. **Sumudan von Dreieich**

Ein Ritter aus Andergast, der ein Gemälde des Neunaugensees anfertigen soll.
Er sucht nach einem besonders schönen Ort am See und bittet Nindariel um Hilfe.

3. **Magister Bergor Brick**

Ein Magier, der den See wegen seiner magischen Geheimnisse erforschen will.
Er interessiert sich für:

* Kairan (magisches Wasser)

* Legenden über Ungeheuer, Irrlichter, Geister

* die Fee Pandlaril

* den Elfenmagier Sindal Seeträumer, der ihn angeblich erwartet

4. **Der alte Gerol**

Ein Einheimischer, der viel über Pandlaril, Träume und die Schreine am See erzählt.
Er berichtet von:

* Harald von Olat, einem Mann, der oft mit Pandlaril im Traum spricht

* einem Schrein der Fee im Dorf Ufern

* magischen Artefakten im Firun-Tempel (z. B. „Eisige Steine wider das Feuer“)

### Zentrale Themen
* Der See ist magisch, voller Geheimnisse, Träume und Gefahren.

* Pandlaril, die Fee des Wassers, schützt die Menschen – aber nicht immer.

* Cealan, der schwarze Berg, scheint sich zu regen und könnte Feuer speien.

* Die Reisenden haben unterschiedliche Ziele, aber der See verbindet sie.
